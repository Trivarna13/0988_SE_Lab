```python
# Hard-Coding Variables

import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

# Hard-code initial variables
a = 0.1
b = -0.5
c = 1.0

# Generate time values
time_values = np.linspace(0, 51, 100)  # Generating 100 time values from 0 to 51

# Calculate temperature values using the quadratic solution
temperature_values = quadratic_solution(time_values, a, b, c)

plt.plot(time_values, temperature_values)
plt.title('Weather Modeling using Quadratic Solution (Hard-Coding Variables)')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (Celsius)')
plt.grid(True)
plt.show()
```

```python
# Keyboard Input

import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

# keyboard input
a = float(input("Enter a value: "))
b = float(input("Enter b value: "))
c = float(input("Enter c value: "))

# Generate time values
time_values = np.linspace(0, 51, 100)  # Generating 100 time values from 0 to 51

# Calculate temperature values using the quadratic solution
temperature_values = quadratic_solution(time_values, a, b, c)

plt.plot(time_values, temperature_values)
plt.title('Weather Modeling using Quadratic Solution (Keyboard Input)')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (Celsius)')
plt.grid(True)
plt.show()
```

```python
# Read From File

import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

# read from file
def read_coefficients_from_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    return [float(line.strip()) for line in lines]

filename = input("Enter the filename containing coefficients (one per line): ")
coefficients = read_coefficients_from_file(filename)
a, b, c = coefficients
print(f"a = {a}, b = {b}, c = {c}")

# Generate time values
time_values = np.linspace(0, 51, 100)  # Generating 100 time values from 0 to 51

# Calculate temperature values using the quadratic solution
temperature_values = quadratic_solution(time_values, a, b, c)

plt.plot(time_values, temperature_values)
plt.title('Weather Modeling using Quadratic Solution (Read From File)')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (Celsius)')
plt.grid(True)
plt.show()
```

```python
# Multiple Sets of Inputs
import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

# multiple sets of inputs
coefficients = [
    [1.0, 0.2, 2.0],  # Example set 1: a, b, c
    [0.8, 0.5, 1.0],  # Example set 2: a, b, c
    [0.5, -0.8, 0.5],  # Example set 3: a, b, c
    [1.5, 2.0, 1.1]   # Example set 4: a, b, c
]

# Generate time values
time_values = np.linspace(0, 51, 100)  # Generating 100 time values from 0 to 51

for i, (a, b, c) in enumerate(coefficients):
    # Calculate temperature values using the quadratic solution
    temperature_values = quadratic_solution(time_values, a, b, c)
    plt.plot(time_values, temperature_values, label=f'Set {i+1}: a={a}, b={b}, c={c}')

plt.title('Weather Modeling using Quadratic Solution (Multiple Sets of Inputs)')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Hard-coding variables & Keyboard Input
import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

# Hard-code initial variables
a1 = 0.2
b1 = -0.5
c1 = 1.0

# Keyboard input
a2 = float(input("Enter a value: "))
b2 = float(input("Enter b value: "))
c2 = float(input("Enter c value: "))

# Generate time values
time_values = np.linspace(0, 51, 100)  # Generating 100 time values from 0 to 51

# Calculate temperature values using the quadratic solution
temperature_values1 = quadratic_solution(time_values, a1, b1, c1)
temperature_values2 = quadratic_solution(time_values, a2, b2, c2)

plt.plot(time_values, temperature_values1, label=f'Hard-coded: a={a1}, b={b1}, c={c1}')
plt.plot(time_values, temperature_values2, label=f'Keyboard Input: a={a2}, b={b2}, c={c2}')

plt.title('Weather Modeling using Quadratic Solution (Hard-coding variables & Keyboard Input)')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.grid(True)
plt.show()
```
